```python
import pandas as pd

# --- E: EXTRAÇÃO ---
# Lendo o arquivo que criamos
df = pd.read_csv('Desafio_ETL.csv')

# --- T: TRANSFORMAÇÃO ---
def gerar_mensagem(nome, saldo):
    if saldo > 1000:
        return f"Olá {nome}, que tal investir seu saldo em nossa nova carteira premium?"
    else:
        return f"Olá {nome}, confira nossas dicas para poupar e aumentar seu saldo!"

# Aplicando a lógica para criar uma nova coluna
df['Mensagem'] = df.apply(lambda row: gerar_mensagem(row['Nome'], row['Saldo']), axis=1)

# --- L: CARREGAMENTO ---
# Salvando o resultado em um novo arquivo
df.to_csv('usuarios_com_mensagens.csv', index=False)

print("ETL concluído com sucesso! Verifique o arquivo 'usuarios_com_mensagens.csv'.")
```
